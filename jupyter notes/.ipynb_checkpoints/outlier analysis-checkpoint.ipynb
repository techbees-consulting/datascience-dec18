```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=list(range(1,100,2))
y=list(range(1,51))
```

```python
x[10]=300
x[20]=400
x[26]=250
x[40]=500
```

```python
odata=pd.DataFrame({'x':x,'y':y})
```

```python
sns.relplot(x=x,y=y,data=odata)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
sns.boxplot(x=odata['x'])
plt.show()
```

```python
def is_outlier(data):
    q1,q3 = np.percentile(data,[25,75])
    IQR=q3-q1
    lower_bound = q1-(1.5*IQR)
    upper_bound = q3+(1.5*IQR)
    for i in data:
        if i<lower_bound | i>upper_bound:
            return True
        else:
            return False
```

```python
q1,q3 = np.percentile(odata,[25,75])
IQR=q3-q1
lower_bound = q1-(1.5*IQR)
upper_bound = q3+(1.5*IQR)   
```

```python
outliers=odata[(odata['x']>upper_bound) | (odata['x']<lower_bound) ]
sns.relplot(x='y',y='x',data=outliers)
```
