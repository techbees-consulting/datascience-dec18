```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('C:/Users/ak/Desktop/techbees/datasets/iris.csv')
```

```python
iris.shape
iris.info()
```

```python
counts,binedges=np.histogram(iris['sepal_length'],bins=5,density=True)
```

```python
pdf=counts/sum(counts)
```

```python
plt.plot(binedges[1:],pdf)
plt.show()
```

```python
cdf = np.cumsum(pdf)
```

```python
plt.plot(binedges[1:],cdf)
plt.plot(binedges[1:],pdf)
plt.show()
```
