```python
import numpy as np
arr1= np.arange(1,10,0.5)
```

```python
plt.plot(arr1,arr1**2,'r--',linewidth=5,label='^2')
plt.plot(arr1,arr1**3,'go',label='^3')
plt.plot(arr1,arr1**4,'b^',label='^4')
plt.xlabel('index')
plt.ylabel('values')
plt.title('multiple lines')
plt.legend()
plt.show()
```

```python
import math
asin=[math.sin(c) for c in arr1]
atan=[math.tan(c) for c in arr1]
```

```python
plt.figure(1)
plt.subplot(221)
plt.plot(arr1,asin,'b-')
plt.subplot(222)
plt.plot(arr1,atan,'r-')
plt.show()
```
